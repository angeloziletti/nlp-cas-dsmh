## Step 1 — Create your Gemini API key

1. Open **[aistudio.google.com](https://aistudio.google.com)** in your browser.
2. **Sign in** with your Google account.
3. In the **left sidebar**, click **Get API key**.
4. Click **Create API key**.
5. A long string of letters and numbers appears — that is your key. **Copy it.** Keep it for a moment; you will paste it in Step 2.

A few notes:
- The key is **free** and needs no credit card.
- Google AI Studio lets you view the key again later, so you will not lose it.
- If AI Studio shows a banner asking you to **restrict** the key, click **"Restrict to Gemini API"** — it takes one click and keeps your key working long-term.
- Treat the key like a password — do not share it or post it anywhere public.

---
## Step 2 — Store the key safely

You need the key to be readable by Python without ever appearing in a cell. Pick the path that matches how you are running this notebook.

### Step 2A — In Google Colab *(this is what the course uses)*

1. In the **left sidebar of Colab**, click the **🔑 key icon** (hover text: *Secrets*).
2. Click **+ Add new secret**.
3. In the **Name** field, type exactly:

   `GEMINI_API_KEY`

   All capitals, with underscores. **The spelling must match exactly** — this is the single most common setup mistake.
4. In the **Value** field, **paste the key** you copied in Step 1.
5. Turn **on** the **Notebook access** toggle next to the secret.

Your key is now stored privately in Colab. The notebook can read it — but the key itself never appears in any cell.

### Step 2B — Running locally *(optional, only if not in Colab)*

If you have chosen to run the notebooks on your own machine, copy the repo's `.env.example` to `.env` and paste your key in there:

```bash
cp .env.example .env
# then open .env and replace the placeholder with your real key
```

`.env` is gitignored, so the key never leaves your machine. The setup cell below picks it up automatically. (An exported `GEMINI_API_KEY` environment variable also works if you prefer that.)

## Step 3 — Install, connect, and test

Run the three cells below in order (`Shift + Enter`). If they all succeed, your setup is complete.

````python
!pip install -q google-genai
````

````python
import os
from google import genai

# Read the key — Colab Secrets first, with a local .env / environment-variable fallback.
try:
    from google.colab import userdata
    api_key = userdata.get("GEMINI_API_KEY")
except ImportError:
    from dotenv import load_dotenv, find_dotenv
    load_dotenv(find_dotenv())
    api_key = os.environ.get("GEMINI_API_KEY")

if not api_key:
    raise RuntimeError(
        "GEMINI_API_KEY not found. In Colab, store it as a Secret named "
        "GEMINI_API_KEY. Locally, copy .env.example to .env and paste your "
        "key there (or set the environment variable before launching Jupyter)."
    )

client = genai.Client(api_key=api_key)
print("API key loaded and client created.")
````

If the cell above ran without error, your key is stored correctly. *(If it raised an error, see **Troubleshooting** at the bottom.)*

Now the real test — an actual call to the model:

````python
response = client.models.generate_content(
    model="gemini-3.1-flash-lite",
    contents="In one short, friendly sentence, introduce yourself to a student "
             "starting a course on NLP and large language models in healthcare.")
print(response.text)
````

## Troubleshooting

If a cell raised an error, find the closest match below.

| What you see | What to do |
|---|---|
| The **connect cell** errored, mentioning a *secret* or *not found* | The secret name is wrong, or its access is off. Open the **🔑 Secrets** panel and check: the name is **exactly** `GEMINI_API_KEY` (no spaces, no typo), and the **Notebook access** toggle is **on**. Then re-run the cell. |
| The **API call cell** errored, mentioning an *invalid key* | The key was copied incompletely or with a stray space. Re-copy the whole key from AI Studio, update the **Value** in the Secrets panel, and re-run. |
| An error mentioning *quota*, *rate limit*, or *resource exhausted* | The free tier has usage limits. Wait a minute and run the cell again. |
| A banner in AI Studio about an *unrestricted key* | Not an error. Click **"Restrict to Gemini API"** in AI Studio — it keeps your key working. |

After any fix, re-run the cells in Step 3 from the top.
